```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# mol vol.s same for different species

def floryHugginsFreeEnergy(phi, chi):
    res = phi*np.log(phi) + (1-phi)*np.log(1-phi) + chi*phi*(1-phi)
    return res
```
