```python
import numpy as np
DTYPE = np.float64
epsilon = np.finfo(DTYPE).eps

import matplotlib.pyplot as plt
```

```python
def floryHugginsFreeEnergy(phi:DTYPE, chi:DTYPE):
    res = phi*np.log(phi) + (1-phi)*np.log(1-phi) + chi*phi*(1-phi)
    return res
```

```python
'''
System Parameters:
Given: the total concentration of component 1 in the system
'''
PHI_GLOBAL = DTYPE(0.3) # Setting the global value of the first component in the system
ETA_GLOBAL = DTYPE(1) # Setting the global system size as 1


'''
Flory Parameter(s)
'''
chi = DTYPE(2.1) # test chi


'''
Compartment-wise parameters:
eta_c1 -> size of compartment 1
phi_c1 -> concentration of component 1 in compartment 1
'''

eta_c1 = np.linspace(0, 1, 10, dtype = DTYPE)


phi_c1 = np.linspace(epsilon, 1-epsilon, 10, dtype=DTYPE) # to prevent ill-defined logs in FH FE

phi_c2 = np.ones(len(phi_c1), dtype = DTYPE)
phi_c2 -= phi_c1
```
